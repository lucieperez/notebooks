```python
df = pd.read_csv("data/annotation_df_history/isaiah_dataset.csv")
```

```python
# Create a new column to store the comparison results
df['length_comparison'] = ""

# Iterate through the DataFrame rows
for index, row in df.iterrows():
    if 'MT' not in row['scroll']:
        # Calculate the length of the strings in 'gcons_verse' and 'sign_info'
        gcons_verse_length = len(row['gcons_verse'])
        sign_info_length = len(row['sign_info'])

        # Compare the lengths and set the 'length_comparison' column accordingly
        if gcons_verse_length == sign_info_length:
            df.at[index, 'length_comparison'] = True
            print(row)
        else:
            df.at[index, 'length_comparison'] = False
            print(row.gcons_verse, "\n", row.sign_info)
    else:
        df.at[index, 'length_comparison'] = 'Not Applicable'

# Display the DataFrame with the comparison results
#print(df)
```
