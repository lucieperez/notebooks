```python
# Adding spatial argument type annotations
```

```python
df2 = pd.read_csv("data/sbl_grouped_cmpls_for_annot.csv")
```

```python
df2 = df2[(df2.scroll == "MT") & (df2.book == "Exodus")]
```

```python
# Merge the spatial arguments columns from df2 to df

# Create a dictionary to store the data from df2
spatial_arg_type = {}

for i, row in df2.iterrows():
    spatial_arg_type[(row.verb_id, row.complement)] = row.spatial_arg_type
```

```python
# Create a list in which we store the content of the dictionary, and add it to df

new_col = []

for i, row in df.iterrows():
    new_col.append(spatial_arg_type.get((row.verb_id, row.complement), ""))

df["spatial_arg_type"] = new_col
```

```python
# Add "source" for complements with MN (previously excluded)

df.loc[df.preposition_1 == "MN", 'spatial_arg_type'] = "source"
```

```python
df.head(10)
```

```python
# Create a filtered df with occurrences of BW>[ only

filtered_df = (df[df.lex == "BW>["]).copy().reset_index()
```

```python
set(filtered_df.spatial_arg_type)
```

```python
# Rername the filtered_df (easier)
df_hb = filtered_df
df_hb['verb_id'] = df_hb['verb_id'].astype(int).astype(str)
```

```python
set([type(verb_id) for verb_id in df_hb.verb_id])
```

```python
# Upload the LXX dataset
df_lxx = pd.read_csv("data/exodus_lxx.csv")
```

```python
# Visualise the data (2 rows)
df_lxx.head(2)
```

```python
# Handle some differences between the two datasets
df_lxx = df_lxx.rename(columns={"verse":"gcons_verse"})
df_lxx["scroll"] = "LXX"
df_lxx["book"] = "Exodus"
df_lxx.head(5)
```

```python
# Convert chapter and verse_num in df_hb to integers
df_hb['chapter'] = df_hb['chapter'].astype(int)
df_hb['verse_num'] = df_hb['verse_num'].astype(int)

# Convert chapter and verse_num in df_lxx to integers
df_lxx['chapter'] = df_lxx['chapter'].astype(int)
df_lxx['verse_num'] = df_lxx['verse_num'].astype(int)
```

```python
# Merge df_lxx with df_hb on both 'chapter' and 'verse_num'
df_lxx_filtered = pd.merge(df_lxx, df_hb[['chapter', 'verse_num']], on=['chapter', 'verse_num'], how='inner')
```

```python
df_combined = pd.concat([df_hb, df_lxx_filtered]).fillna("").drop(columns="index")
```

```python
df_combined
```

```python
# Sort the data
# Define a custom order for the 'scroll' column (to have MT first, for annotation purposes)
df_combined['scroll'] = pd.Categorical(df_combined['scroll'], categories=['MT', 'LXX'], ordered=True)

# Sort the value
df_combined = df_combined.sort_values(by=['chapter', 'verse_num', 'scroll'])
```

```python
df_combined.head(4)
```

```python
set([type(verb_id) for verb_id in df_combined.verb_id])
```

```python
df = df_combined.copy().reset_index(drop=True)
```

```python
hb_chapter_verse = {}
lxx_chapter_verse = {}
lxx_occ = {}

for i, row in df.iterrows():
    key = (row.chapter, row.verse_num)
    if row.scroll == "MT":
        hb_chapter_verse.setdefault(key, []).append(i)
    else:
        lxx_chapter_verse.setdefault(key, []).append(i)
        lxx_occ[i] = len(lxx_chapter_verse[key]) - 1
    
for i, row in df.iterrows():
    if row.scroll == "MT":
        df.at[i, "verb_id"] = df.at[i, "verb_id"] + "_mt"
    else:
        mt_index = hb_chapter_verse[(row.chapter, row.verse_num)][lxx_occ[i]]
        df.at[i, "verb_id"] = df.at[mt_index, "verb_id"].replace("_mt", "_lxx")
```
