```python
import pandas as pd
```

```python
df1 = pd.read_excel("data/new_dataset_mt_lxx_data.xlsx").fillna("")
```

```python
df2 = pd.read_csv("data/exodus_bo_dss.csv").fillna("")
```

```python
df2 = df2[df2.comments != "reconstructed"].copy().reset_index()
```

```python
frames = [df1, df2]
```

```python
df = pd.concat(frames).fillna("")
```

```python
df = df.reset_index()
```
