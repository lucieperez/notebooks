```python
# Add a new variable: construction (prepositional vs non prepositional, prep / non_prep)
set(df.gc)

# Create a mapping dictionary
mapping = {
    'hey': 'non-prep',
    'acc of direction': 'non-prep',
    'prep': 'prep',
    'prep and prep': 'prep'
}

# Create the new column 'construction' using the map function
df['construction'] = df['gc'].map(mapping)

# Verification
print("Set of gc for prep construction: ", set(df[df.construction == "prep"].gc))
print("Set of gc for non-prep construction: ", set(df[df.construction == "non-prep"].gc))
```

## 3. Chi Square Test of Independence: Cmpl Construction and Cmpl Definiteness

```python
# Contingency table
contingency_table_det = pd.crosstab(df['construction'], df['gc_def'])

# Convert the contingency table to a numpy array
observed = contingency_table_det.values

# Perform the chi-square test
chi2, p, dof, expected = chi2_contingency(observed)

# Output the results
print("Chi-Square Statistic:", chi2)
print("P-value:", p)
print("Degrees of Freedom:", dof)
print("Expected Frequencies: the expected frequencies should all be >= 5 for the Chi2 test to be appropriate\n", expected)
```

```python
# Medill
contingency_table_det = contingency_table_det.T
contingency_table_det
```

```python
contingency_table_det.to_csv("data/sbl_medill_with_sham.csv")
```

## 4. Construction2 (prep, dir-he, vc)

```python
# Add a new variable: construction (prepositional vs non prepositional, prep / non_prep)
set(df.gc)

# Create a mapping dictionary
mapping_2 = {
    'hey': 'dir-he',
    'acc of direction': 'vc',
    'prep': 'prep',
    'prep and prep': 'prep'
}

# Create the new column 'construction' using the map function
df['construction_2'] = df['gc'].map(mapping_2)

# Verification
print("Set of cmpl_constr for prep construction: ", set(df[df.construction_2 == "prep"].gc))
print("Set of cmpl_constr for vc construction: ", set(df[df.construction_2 == "vc"].gc))
print("Set of cmpl_constr for dir-he construction: ", set(df[df.construction_2 == "dir-he"].gc))
```

## 5. Chi Square Test of Independence: Construction 2 and Definiteness

```python
# Contingency table
contingency_table_det = pd.crosstab(df['construction_2'], df['gc_def'])

# Convert the contingency table to a numpy array
observed = contingency_table_det.values

# Perform the chi-square test
chi2, p, dof, expected = chi2_contingency(observed)

# Output the results
print("Chi-Square Statistic:", chi2)
print("P-value:", p)
print("Degrees of Freedom:", dof)
print("Expected Frequencies: the expected frequencies should all be >= 5 for the Chi2 test to be appropriate\n", expected)
```
