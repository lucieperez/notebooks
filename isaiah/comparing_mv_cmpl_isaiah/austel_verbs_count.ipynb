```python
import pandas as pd

from itertools import chain

from tfob import TFOb, BHSA, DSS
```

```python
austel_verbs = ['BW>[','HLK[','JY>[','JRD[','<BR[',
                '<LH[','CWB[','>TH[','BRX[','GJX[',
                'GLH[','GLL[','DXP[','DLG[','HWH[',
                'XWC[','XLP[','XSH[','VB<[','VWF[',
                'MHR[','MWC[','NGC[','NHR[','NWX[',
                'NWS[','NXT[','NVP[','NS<[','NPL[',
                'NTK[','SBB[','SWR[','SLQ[','<WZ[',
                '<WP[','PNH[','PC<[','YWP[','Y<D[',
                'QHL[','QPY[','QRB[','RWY[','FVH[',
                'CWX[','CWR[','CVP[','CQQ[','T>R[',
                'T<H[']
```

```python
len(austel_verbs)
```

```python
verb_counts = []

for verb in austel_verbs:
    verb_count = len(TFOb.all("word", BHSA).filter(lex=verb))
    verb_counts.append(verb_count)
```

```python
df = pd.DataFrame({"verb": austel_verbs, "verb_count": verb_counts})
df = df.sort_values(by=["verb_count"], ascending=False, ignore_index=True)
df
```
