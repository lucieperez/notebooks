```python
from tfob import TFOb, BHSA, DSS
```

```python
import pandas as pd
```

```python
isaiah_df = pd.read_csv("data/isaiah_cl_types.csv")
```

```python
isaiah_df
```

```python
verb_ids = list(isaiah_df["verb_id"])
```

```python
questions = []

for verb_id in verb_ids:
    verse = TFOb(verb_id, BHSA).to_verses
    clauses = verse.to_clauses
    question = f"(id {verb_id}) The verse is {verse.text}. \n"

    for clause in clauses:
        question += f" {clause.text} is a clause. \n"
    questions.append(question)
```
