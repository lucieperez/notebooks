```python
import pandas as pd
```

```python
df = pd.read_csv("data/annotated_isaiah_dataset.csv", sep=";")
```

```python
df
```

```python
df["chapter"] = pd.to_numeric(df["chapter"])
df["verse_num"] = pd.to_numeric(df["verse_num"])
# converting 'dir_he' and 'dir_he_dss' from float to int 

df["dir_he"] = df["dir_he"].astype(str)
df["dir_he"] = df["dir_he"].replace("1.0", "1").replace("0.0", "0")

df["dir_he_dss"] = df["dir_he_dss"].astype(str)
df["dir_he_dss"] = df["dir_he_dss"].replace("1.0", "1").replace("0.0", "0")

df.fillna("")
```

```python
# Dataset tests 

# Find lines without complement and without "no complement"
df[(df.complement == "")]
```

```python
df[(df.preposition_1 == ' >L')]
```

```python
# Find lines with complement but missing dir_he data
df[(df.complement != "") & (df.dir_he == "")]
```

```python
# Checking the list of prepositions

prepositions1 = df['preposition_1'].tolist()
prepositions2 = df['preposition_2'].tolist()
prepositions3 = df['preposition_3'].tolist()
prepositions4 = df['preposition_4'].tolist()

prepositions = prepositions1 + prepositions2 + prepositions3 + prepositions4

set(prepositions)
```

```python
# Checking the list of verbs

verbs = set(df["lex"].tolist())
verbs
```

```python
# Create a new dataset with verbs having complements only

df_cmpl = df

df_cmpl.drop(df_cmpl[df_cmpl['complement'] == "no complement"].index, inplace = True)
```

```python
df_cmpl.fillna("")

# Remove the dir_he_dss column

df_cmpl.drop("dir_he_dss", axis="columns").fillna("")
```

```python
df_cmpl.to_csv("data/isaiah.csv", sep=",", index=False)
```
