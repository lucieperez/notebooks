```python
df = pd.read_csv("DATASET.csv").fillna("")
```

```python
df2 = pd.read_csv("DATASET2.csv")
```

```python
rows_by_verb_id = {}
for i, row in df2.iterrows():
    if row.verb_id != "s":
        rows_by_verb_id.setdefault(int(row.verb_id), []).append(row)
```

```python
verb_ids = []
for verb_id, row in rows_by_verb_id.items():
    if not df[df.verb_id == verb_id].iloc[0].complement and row.complement:
        verb_ids.append(verb_id)
```

```python
items = []
for i, row in df.iterrows():
    item = row.to_dict()
    if row.complement:
        items.append(item)
        continue
        
    verb_id = row.verb_id
    hand_filled_rows = rows_by_verb_id.get(verb_id)
    if not hand_filled_rows:
        items.append(item)
    else:
        for hand_filled_row in hand_filled_rows:
            item = hand_filled_row.to_dict()
            item["verb_id"] = int(item["verb_id"])
            items.append(item)

merged = pd.DataFrame(items)
merged
```
