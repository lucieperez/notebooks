```python
import pandas as pd
```

### Create a dataframe from the verb ids to compare GPT's response to the ETCBC analysis

```python
df = pd.read_csv("data/sample_df1.csv")
```

```python
df_answers = pd.read_csv("data/gpt_answers_sample-df1.csv")
df_answers = df_answers.drop(columns="Unnamed: 0")
df_answers
```

```python
# Get the clauses from the ETCBC

clauses = []

for _, row in df.iterrows():
    verb_id = row.verb_id
    clause = TFOb(verb_id, BHSA).to_clauses.text
    clauses.append(clause)
```

```python
# Add the right clauses to the df_answer
df_answers["ETCBC_clause"] = clauses
df_answers
```

```python
# Save the dataframe
df_answers.to_csv("data/check_clauses_sample_df1.csv")
```
