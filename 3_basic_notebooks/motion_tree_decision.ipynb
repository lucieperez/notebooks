```python
from graphviz import Digraph
```

```python
dot = Digraph("motion_tree", format="svg")  # or "pdf"
dot.attr(rankdir="TB", fontsize="12")
dot.attr("node", shape="box", style="rounded", fontsize="11")
dot.attr("edge", fontsize="10")
```

```python
# Nodes
dot.node("N1", "Node 1: Path/Goal configuration present or inferable?")
dot.node("A", "Exit A: Not motion", shape="oval", style="filled", fillcolor="lightgray")

dot.node("N2", "Node 2: Actual displacement of a Figure relative to external frame?")
dot.node("B", "Exit B: FACTIVE motion", shape="oval", style="filled", fillcolor="lightblue")

dot.node("N3", "Node 3: Dynamic scanning / imagined motion in cognition?")
dot.node("C", "Exit C: Not motion\n(change-of-state / abstract metaphor)", shape="oval", style="filled", fillcolor="lightgray")
dot.node("D", "Exit D: FICTIVE motion -> subtype", shape="oval", style="filled", fillcolor="lightyellow")

dot.node("N4", "Subtype: Stationary extended entity traversed?\n(road/border/river/wall...)")
dot.node("S1", "FICTIVE: Coextension path", shape="oval", style="filled", fillcolor="lightyellow")

dot.node("N5", "Subtype: Orientation implies vector?\n(face/point toward...)")
dot.node("S2", "FICTIVE: Orientation path", shape="oval", style="filled", fillcolor="lightyellow")

dot.node("N6", "Subtype: Emanation/transmission?\n(sound/light/gaze/odor)")
dot.node("S3", "FICTIVE: Emanation (radiation) path", shape="oval", style="filled", fillcolor="lightyellow")

dot.node("N7", "Subtype: Observer-based apparent motion?")
dot.node("S4", "FICTIVE: Observer-based", shape="oval", style="filled", fillcolor="lightyellow")

dot.node("N8", "Subtype: Abstract Figure, no spatial support?")
dot.node("S5", "Metaphorical motion\n(or cline: high metaphoricity)", shape="oval", style="filled", fillcolor="wheat")

# Edges
dot.edge("N1", "A", label="No")
dot.edge("N1", "N2", label="Yes")

dot.edge("N2", "B", label="Yes")
dot.edge("N2", "N3", label="No")

dot.edge("N3", "C", label="No")
dot.edge("N3", "N4", label="Yes")

dot.edge("N4", "S1", label="Yes")
dot.edge("N4", "N5", label="No")

dot.edge("N5", "S2", label="Yes")
dot.edge("N5", "N6", label="No")

dot.edge("N6", "S3", label="Yes")
dot.edge("N6", "N7", label="No")

dot.edge("N7", "S4", label="Yes")
dot.edge("N7", "N8", label="No")

dot.edge("N8", "S5", label="Yes")
dot.edge("N8", "D", label="No / other")
```

```python
# Render
dot.render("motion_decision_tree", cleanup=True)
```
