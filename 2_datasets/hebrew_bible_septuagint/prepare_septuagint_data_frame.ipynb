```python
# Adding spatial argument type annotations
```

```python
df2 = pd.read_csv("data/sbl_grouped_cmpls_for_annot.csv")
```

```python
df2 = df2[(df2.scroll == "MT") & (df2.book == "Exodus")]
```

```python
# Merge the spatial arguments columns from df2 to df

# Create a dictionary to store the data from df2
spatial_arg_type = {}

for i, row in df2.iterrows():
    spatial_arg_type[(row.verb_id, row.complement)] = row.spatial_arg_type
```

```python
# Create a list in which we store the content of the dictionary, and add it to df

new_col = []

for i, row in df.iterrows():
    new_col.append(spatial_arg_type.get((row.verb_id, row.complement), ""))

df["spatial_arg_type"] = new_col
```

```python
# Add "source" for complements with MN (previously excluded)

df.loc[df.preposition_1 == "MN", 'spatial_arg_type'] = "source"
```

```python
df.head(10)
```

```python
# Create a filtered df with occurrences of BW>[ only

filtered_df = (df[df.lex == "BW>["]).copy().reset_index()
```

```python
set(filtered_df.spatial_arg_type)
```

```python
# Adding LXX rows
```

```python
# Create a list to hold the new dataframe rows
new_rows = []

# List of columns that I want to keep empty
empty_columns = ['lex', 'gcons_verb', 'gcons_verse', 'stem', 'tense', 'gcons_clause', 'subject',
       'complement', 'cmpl_lex', 'cmpl_translation', 'dir_he', 'cmpl_constr',
       'cmpl_det', 'cmpl_indiv', 'cmpl_complex', 'motion_type',
       'spatial_arg_type', 'preposition_1', 'preposition_2', 'preposition_3',
       'preposition_4', 'comments']
```

```python
# Iterate over each row in the original DataFrame
for index, row in filtered_df.iterrows():
    # Append the original row to the list
    new_rows.append(row)
    
    # Create the new row with specific modifications
    new_row = row.copy()  # Copy the original row
    
    # For specific columns, retain information from the original row
    new_row['verb_id'] = row['verb_id']
    new_row['book'] = row['book']
    new_row['chapter'] = row['chapter']
    new_row['verse_num'] = row['verse_num']
    
    # For another specific column, add new information
    new_row['scroll'] = "LXX"
    
    # For columns that should remain empty in the new row
    for col in empty_columns:
        new_row[col] = ""

    # Append the modified row (new row) to the list
    new_rows.append(new_row)

# Create a new DataFrame with the list of rows (original + new)
new_df = pd.DataFrame(new_rows)

# Reset the index (optional)
new_df = new_df.reset_index(drop=True)
```

```python
new_df = new_df.drop("index", axis=1)
```
