```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df_mt = pd.read_csv("../generate_datasets/data/bhsa_all_verbs.csv").fillna("")
df_xb = pd.read_csv("../generate_datasets/data/extrabiblical_all_verbs.csv").fillna("")
df_bib_dss = pd.read_csv("../generate_datasets/data/bib_dss_all_verbs.csv").fillna("")
```

```python
# converting 'dir_he' and 'dir_he_dss' from float to int 

df_mt["dir_he"] = df_mt["dir_he"].astype(str)
df_mt["dir_he"] = df_mt["dir_he"].replace("1.0", "1").replace("0.0", "0")

df_xb["dir_he"] = df_xb["dir_he"].astype(str)
df_xb["dir_he"] = df_xb["dir_he"].replace("1.0", "1").replace("0.0", "0")

df_bib_dss["dir_he"] = df_bib_dss["dir_he"].astype(str)
df_bib_dss["dir_he"] = df_bib_dss["dir_he"].replace("1.0", "1").replace("0.0", "0")
```

```python
#df_mt.head(5)
#df_bib_dss.head(5)
#df_xb.head(5)
```

```python
#Make dictionaries of verb counts

mt_lex = df_mt["lex"].value_counts()
xb_lex = df_xb["lex"].value_counts()
bib_dss_lex = df_bib_dss["lex"].value_counts()
```

```python
mt_lex
```

```python
# Plot the number of occurrences of each verb in BHSA

keys = list(mt_lex.keys())
vals = [mt_lex[k] for k in keys]


plt.figure(figsize=(20,14))
sns.barplot(x=vals, y=keys)
```

```python
# Plot the number of occurrences of each verb in Biblical Dead Sea Scrolls

keys = list(bib_dss_lex.keys())
vals = [bib_dss_lex[k] for k in keys]


plt.figure(figsize=(20,14))
sns.barplot(x=vals, y=keys).set(title='Occurrences of Motion Verbs in the Masoretic Text')
```

```python
# Plot the number of occurrences of each verb in XB

keys = list(xb_lex.keys())
vals = [xb_lex[k] for k in keys]


plt.figure(figsize=(20,14))
sns.barplot(x=vals, y=keys).set(title='Occurrences of Motion Verbs in Non Biblical Dead Sea Scrolls')
```
