```python
import pandas as pd
```

```python
df1 = pd.read_csv("data/bhsa_all_verbs_without_min.csv").fillna("")
len(df1)
```

```python
df2 = pd.read_csv("data/bib_dss_all_verbs_without_min.csv").fillna("")
len(df2)
```

```python
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)
```

```python
df1.columns
```

```python
df2.columns
```

```python
frames = [df1, df2]
```

```python
df = pd.concat(frames).fillna("")
```

```python
len(df)
```

```python
df.sort_values(["book", "chapter", "verse_num"], ascending=[True, True, True], ignore_index=True, inplace=True)
len(df)
```

```python
# Add annotation columns: 'cmpl_translation', 'motion_type', 'comments'
df['cmpl_translation'] = ""
df['motion_type'] = "" 
df['comments'] = ""
```

```python
len(df)
df.columns
```

```python
df.head(10)
```

```python
# Create a custom sorting function for the 'scroll' column where 'MT' comes first
def prioritize_mt(scroll):
    if scroll == 'MT':
        return 0
    else:
        return scroll



# Apply the function to create a new 'sort_key' column
df['sort_key'] = df['scroll'].apply(prioritize_mt)

# Convert 'sort_key' to Categorical, taking into account that non-MT values should be sorted alphabetically
df['sort_key'] = pd.Categorical(df['sort_key'], categories=[0] + sorted(df[df['scroll'] != 'MT']['scroll'].unique()), ordered=True)

# Sort the DataFrame by the necessary columns
df.sort_values(["book", "chapter", "verse_num", "sort_key"], ascending=[True, True, True, True], ignore_index=True, inplace=True)

# Optional: Remove the 'sort_key' column if it's no longer needed
df.drop('sort_key', axis=1, inplace=True)
```
