```python
import pandas as pd
```

```python
df1 = pd.read_csv("data/bhsa_all_verbs.csv").fillna("")
```

```python
df2 = pd.read_csv("data/bib_dss_all_verbs.csv").fillna("")
```

```python
df1.columns
```

```python
df2.columns
```

```python
frames = [df1, df2]
```

```python
df = pd.concat(frames).fillna("")
```

```python
df.sort_values(["book", "chapter", "verse_num"], ascending=[True, True, True], ignore_index=True, inplace=True)
df
```

```python
df.to_csv("data/all_verbs_bhsa_bib_dss.csv", index=False)
```
