```python
import pandas as pd
```

```python
df = pd.read_csv("data/isaiah_dataset.csv")
```

```python
pd.set_option('display.max_columns', None)
```

```python
# Getting read of NaN and float numbers

df["chapter"] = pd.to_numeric(df["chapter"])
df["verse_num"] = pd.to_numeric(df["verse_num"])
# converting 'dir_he' and 'dir_he_dss' from float to int 

df["dir_he"] = df["dir_he"].astype(str)
df["dir_he"] = df["dir_he"].replace("1.0", "1").replace("0.0", "0")

df = df.fillna("")
```

```python
# Selecting only the occurrences of motion verb + complement in the dataset
df_cmpl = df[df.complement != "no complement"]
df_cmpl.head(5)
```

```python
# Check how many motion_type are no yet fully annotated
len(df[df.motion_type == "factive fictive"])
```

```python
# Check how many cmpl_anim are no yet fully annotated
len(df[df.cmpl_anim == "anim inanim"])
```

```python
# Check the content of the columns: list and function

columns_to_check = ["dir_he",
                    "cmpl_constr", 
                    "cmpl_nt", 
                    "cmpl_anim",
                    "cmpl_det",
                    "cmpl_indiv",
                    "cmpl_complex",
                    "motion_type",
                    "preposition_1",
                    "preposition_2",
                    "preposition_3",
                    "preposition_4",
                   ]

def show_set_colum_values(df, column_name):
    print(f"In the column {column_name} you have the following values: {set(df[column_name])}.\n")
```

```python
# Check the content of the columns

for column_name in columns_to_check:
    show_set_colum_values(df_cmpl, column_name)
```
